```python
import numpy as np
```

```python
#function for input
def ipt(ip, target):
    temp = []
    weights = ip@target.T
    t =weights.T@ip

    for i in (t):
            if i > 0:
                temp.append(1)
            elif i < 0:
                temp.append(-1)
            else :
                temp.append(0)

    return temp

def opt(ip, target):
    temp = []
    weights = ip@target.T    #6x4
    t = weights@target

    for i in(t):
        if i > 0 :
            temp.append(1)
        elif i < 0:
            temp.append(-1)
        else:
            temp.append(0)

    return temp

ip = np.array([1,0,-1,1,-1,0]).reshape(6,1)  #   #array of postive and negatives
target = np.array([-1,0,1,0]).reshape(4,1)
print("Value is : ", ipt(ip, target))

print("----------------------------------")
print("Value is : ", opt(ip, target))
```
