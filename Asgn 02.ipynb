```python
import numpy as np
```

```python
def and_not (x1, x2):
  w1 = 1
  w2 = -1

  result = []
  for i in range(len(x1)):
    sum = x1[i]*w1 + x2[i]*w2
    if sum >= 0:
      result.append(1)
    else:
      result.append(0)

  return result

x1 = np.array([1, 1, 0, 0])
x2 = np.array([1, 0, 1, 0])

print("Result is : ", )
and_not(x1, x2)
```
