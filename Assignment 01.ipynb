```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def sigmoid(x):
    return 1/ (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0,x)

def leaky_relu (x, alpha = 0.2):
    return np.maximum(alpha*x, x)
def menu():
    print("-----------------------")
    print("1. Sigmoid Function ")
    print("2. Tanh Function ")
    print("3. ReLU Function ")
    print("4. Leaky ReLU Function ")

    ans = input("Choice: ")
    if ans == '1':
        plt.plot(x, sigmoid(x), label='Sigmoid')
    elif ans == '2':
        plt.plot(x, tanh(x), label='Tanh')
    elif ans == '3':
        plt.plot(x, relu(x), label='ReLU')
    elif ans == '4':
        plt.plot(x, leaky_relu(x), label='Leaky_ReLU')
    else:
        print("Invalid choice!")

#defining x-axis values
x = np.linspace(-5, 5, 100)
menu()

plt.xlabel('x')
plt.ylabel('y')
plt.title('Activation Functions')

#main
plt.grid(True)
```
